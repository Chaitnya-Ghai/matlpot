```python
import numpy as np

x=np.linspace(-np.pi , np.pi , 15)
c=np.cos(x)
s=np.sin(x)

# [left , bottom , width , height]
ax=plt.axes([0.1,0.1,0.8,0.8])
# bs: mention blue color, square
# marker with dotted line
ax1 = ax.plot(x,c,"bs:")
# ro- mentions red color,circle
# marker with solid line
ax2= ax.plot(x,s,'ro-')
ax.legend(
    labels = ('cosine Function',
             'Sine Function'),
    loc="upper left"
)
ax.set_title("Trignometric Function")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 15)
c = np.cos(x)
s = np.sin(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot cosine function with blue squares and dotted line
ax1, = ax.plot(x, c, "bs:", label="Cosine Function")

# Plot sine function with red circles and solid line
ax2, = ax.plot(x, s, "ro-", label="Sine Function")

# Add legend
ax.legend(loc="upper left")

# Set title
ax.set_title("Trigonometric Functions")

# Show the plot
plt.show()
```

```python
''' 
Why Use plt.subplots() Instead of plt.axes([])?
More Control: It gives better flexibility when creating multiple plots in the same figure.
Avoids Issues: plt.axes([0.1,0.1,0.8,0.8]) 
creates an axes manually,
which can sometimes behave unexpectedly.
'''
# np.linspace(-np.pi, np.pi, 15) 
# ka kaam yeh hai ki yeh ek array banaata hai jo,
# # -π se π tak 15 evenly spaced values contain karta hai.

# np.linspace(start, stop, num)
# start = -np.pi → Shuru -π se hoga.
# stop = np.pi → End π pe hoga.
# num = 15 → Total 15 values honi chahiye.
```
